```python
group=merged_data.groupby(by=["Sector"]).mean()
#Grouping by Sector
```

```python
group.loc[group['Price']>=100,'Price above 100']=1
group.loc[group['Price']<100,'Price above 100']=0
#Conditional Statements
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```
